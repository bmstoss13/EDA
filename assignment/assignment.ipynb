# Answers to Statistical Properties of Mean and Covariance

## **Q1.** Sample Mean and Covariance Properties

### **1. Show that** $ m(a + bX) = a + b \times m(X) $.

Starting with the definition of the sample mean:
$
m(X) = \frac{1}{N} \sum_{i=1}^{N} x_i
$

Now, consider the transformed variable $ a + bX $:

$
m(a + bX) = \frac{1}{N} \sum_{i=1}^{N} (a + bx_i)
$

Using the linearity of summation,

$
m(a + bX) = \frac{1}{N} \sum_{i=1}^{N} a + \frac{1}{N} \sum_{i=1}^{N} bx_i
$

$
= a + b \times \frac{1}{N} \sum_{i=1}^{N} x_i
$

$
= a + b \times m(X).
$

Thus, we have shown that:

$
m(a + bX) = a + b \times m(X).
$

---

### **2. Show that** $ \text{cov}(X,a+bY) = b \times \text{cov}(X,Y) $.

By definition, the covariance between $ X $ and $ Y $ is:

$
\text{cov}(X,Y) = \frac{1}{N} \sum_{i=1}^{N} (x_i - m(X))(y_i - m(Y)).
$

Consider the transformed variable $ a + bY $:

$
\text{cov}(X, a + bY) = \frac{1}{N} \sum_{i=1}^{N} (x_i - m(X))((a + b y_i) - m(a + bY)).
$

Using the result from part (1),

$
m(a + bY) = a + b m(Y).
$

Substituting this,

$
\text{cov}(X, a + bY) = \frac{1}{N} \sum_{i=1}^{N} (x_i - m(X))(a + b y_i - (a + b m(Y))).
$

Since the constant $ a $ cancels out,

$
\text{cov}(X, a + bY) = \frac{1}{N} \sum_{i=1}^{N} (x_i - m(X))(b(y_i - m(Y))).
$

Factoring out $ b $,

$
\text{cov}(X, a + bY) = b \times \frac{1}{N} \sum_{i=1}^{N} (x_i - m(X))(y_i - m(Y)).
$

$
= b \times \text{cov}(X,Y).
$

Thus, we have shown that:

$
\text{cov}(X,a+bY) = b \times \text{cov}(X,Y).
$

---

### **3. Show that** $ \text{cov}(a+bX,a+bX) = b^2 \text{cov}(X,X) $, and in particular that $ \text{cov}(X,X) = s^2 $.

Using the definition of covariance,

$
\text{cov}(X, X) = \frac{1}{N} \sum_{i=1}^{N} (x_i - m(X))(x_i - m(X)).
$

$
= \frac{1}{N} \sum_{i=1}^{N} (x_i - m(X))^2.
$

By the definition of variance,

$
s^2 = \frac{1}{N} \sum_{i=1}^{N} (x_i - m(X))^2.
$

Thus, we conclude that:

$
\text{cov}(X, X) = s^2.
$

Now, consider the transformation $ a + bX $:

$
\text{cov}(a+bX, a+bX) = b^2 \text{cov}(X, X).
$

$
= b^2 s^2.
$

This confirms the desired result.

---

### **4. Non-Decreasing Transformations and the Median**

The median of $ X $ is the value that splits the dataset into two equal halves. If $ g(X) $ is a non-decreasing transformation, then:

$
\text{Median}(g(X)) = g(\text{Median}(X)).
$

This is because non-decreasing transformations preserve the relative order of values, meaning that the median remains in the same relative position after transformation.

However, this does **not** necessarily hold for other quantiles, such as the interquartile range (IQR) or the range. While quantiles like the median are order-based and preserved under non-decreasing transformations, the **difference-based** statistics (e.g., IQR = $ Q_3 - Q_1 $ and range = max - min) **are not necessarily preserved**.

For example, if $ g(X) = X^2 $, the transformation affects the spread differently depending on whether values are positive or negative.

---

### **5. Is it always true that** $ m(g(X))= g(m(X)) $?

No, it is not always true that:

$
m(g(X)) = g(m(X)).
$

The mean is an expectation operator, whereas $ g $ may introduce non-linearity. The only case where this holds exactly is when $ g(X) $ is a linear function, i.e., $ g(X) = a + bX $, which we proved in part (1).

For general non-linear functions $ g(X) $, the expectation does not commute with $ g $, meaning:

$
m(g(X)) \neq g(m(X)).
$

This follows from Jensenâ€™s inequality: if $ g $ is convex, then:

$
m(g(X)) \geq g(m(X)).
$

Similarly, if $ g $ is concave:

$
m(g(X)) \leq g(m(X)).
$

For example, if $ X $ is normally distributed and $ g(X) = X^2 $, then:

$
m(X^2) = \text{Var}(X) + (m(X))^2 \neq (m(X))^2.
$

Thus, in general:

$
m(g(X)) \neq g(m(X)).
$

In particular, from which countries and giftors are the gifts coming from, and to which institutions are they going? For this question, `.groupby([vars]).count()` and `.groupby([vars]).sum()` will be especially useful to tally the number of occurrences and sum the values of those occurrences.

1. Load the `./data/ForeignGifts_edu.csv` dataset.
2. For `Foreign Gift Amount`, create a histogram and describe the variable. Describe your findings.
3. For `Gift Type`, create a histogram or value counts table. What proportion of the gifts are contracts, real estate, and monetary gifts?

```python
#1.

import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#load dataset
df = pd.read_csv(r"C:\Users\Brian\EDA\assignment\data\ForeignGifts_edu.csv")

#view dataset
print(df.head())
```

```python
#2

#view statistics
print(df["Foreign Gift Amount"].describe())

# Histogram structuring
plt.figure(figsize=(9,5))
sns.histplot(df["Foreign Gift Amount"], bins=50,kde=True)
plt.xlabel("Foreign Gift Amount")
plt.ylabel("Frequency")
plt.title("Distribution of Foreign Gift Amount")
plt.show()
```

The distribution is right skewed - most foreign gifts are relatively small, but a few are very large.

```python
#3

# Count and proportion of each gift type
gift_type_counts = df["Gift Type"].value_counts(normalize=True)
print(gift_type_counts)

# Bar plot structuring
plt.figure(figsize=(8, 5))
sns.barplot(x=gift_type_counts.index, y=gift_type_counts.values)
plt.xlabel("Gift Type")
plt.ylabel("Proportion")
plt.title("Proportion of Each Gift Type")
plt.xticks(rotation=45)
plt.show()
```

Contracts are the most frequent type of gift.

```python
#4

#transform Foreign Gift Amount to log scale
df["Log Foreign Gift Amount"] = np.log1p(df["Foreign Gift Amount"])

# Overall Kernel Density Plot
plt.figure(figsize=(8, 5))
sns.kdeplot(df["Log Foreign Gift Amount"], shade=True)
plt.xlabel("Log Foreign Gift Amount")
plt.title("Kernel Density Plot of Log Foreign Gift Amount")
plt.show()
```

```python
#conditional on gift type
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x="Log Foreign Gift Amount", hue="Gift Type", shade=True)
plt.xlabel("Log Foreign Gift Amount")
plt.title("Kernel Density Plot of Log Foreign Gift Amount by Gift Type")
plt.show()
```

The log transformation reduces skewness. Also, different gift types have different distributions. Contract is larger than monetaary and real estate. 

```python
#5

#count of gifts per country (top 15, as shown by nlargest function)
top_countries_count = df.groupby("Country of Giftor").size().nlargest(15)
print(top_countries_count)

# Total amount given per country (top 15)
top_countries_amount = df.groupby("Country of Giftor")["Foreign Gift Amount"].sum().nlargest(15)
print(top_countries_amount)
```

Certain countries may contribute many small gifts, while others contribute fewer but larger gifts.

```python
#6

#top 15 institutions in terms of the total amount of money they receive
top_institutions = df.groupby("Institution Name")["Foreign Gift Amount"].sum().nlargest(15)
print(top_institutions)

# Histogram of total amount received by institutions
plt.figure(figsize=(8, 5))
sns.histplot(df.groupby("Institution Name")["Foreign Gift Amount"].sum(), bins=50, kde=True)
plt.xlabel("Total Amount Received")
plt.ylabel("Frequency")
plt.title("Distribution of Total Foreign Gifts Received by Institutions")
plt.show()
```

A small number of institutions receive the majority of funds.

```python
#7

#list of the top 15 giftors
top_giftors = df.groupby("Giftor Name")["Foreign Gift Amount"].sum().nlargest(15)
print(top_giftors)
```

Certain giftors contribute disproportionately high amounts.
